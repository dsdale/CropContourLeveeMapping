```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import auc
```

```python
X_norm = np.load ('Graphs/AUCarrays/X_norm.npy')
Y_norm=np.load('Graphs/AUCarrays/Y_norm.npy')

X_RE_2=np.load('Graphs/AUCarrays/X_RE_2.npy')
Y_RE_2=np.load('Graphs/AUCarrays/Y_RE_2.npy')

X_RE_3=np.load('Graphs/AUCarrays/X_RE_3.npy')
Y_RE_3=np.load('Graphs/AUCarrays/Y_RE_3.npy')

X_RE_4=np.load('Graphs/AUCarrays/X_RE_4.npy')
Y_RE_4=np.load('Graphs/AUCarrays/Y_RE_4.npy')

X_RE_5=np.load('Graphs/AUCarrays/X_RE_5.npy')
Y_RE_5=np.load('Graphs/AUCarrays/Y_RE_5.npy')

X_cloud_raw=np.load('Graphs/AUCarrays/X_cloud.npy')
Y_cloud_raw=np.load('Graphs/AUCarrays/Y_cloud.npy')

X_cloud2_raw=np.load('Graphs/AUCarrays/X_cloud2.npy')
Y_cloud2_raw=np.load('Graphs/AUCarrays/Y_cloud2.npy')

X_SENSE_raw=np.load('Graphs/AUCarrays/X_SENSE.npy')
Y_SENSE_raw=np.load('Graphs/AUCarrays/Y_SENSE.npy')

X_SENSE2_raw=np.load('Graphs/AUCarrays/X_SENSE2.npy')
Y_SENSE2_raw=np.load('Graphs/AUCarrays/Y_SENSE2.npy')

X_DIFF=np.load('Graphs/AUCarrays/X_DIFF.npy')
Y_DIFF=np.load('Graphs/AUCarrays/Y_DIFF.npy')

X_CountyMax=np.load('Graphs/AUCarrays/X_CountyMax.npy')
Y_CountyMax=np.load('Graphs/AUCarrays/Y_CountyMax.npy')

X_CountyMin=np.load('Graphs/AUCarrays/X_CountyMin.npy')
Y_CountyMin=np.load('Graphs/AUCarrays/Y_CountyMin.npy')
```

```python
import random 
minEntries = min([X_cloud_raw.shape[0],Y_cloud_raw.shape[0],Y_cloud2_raw.shape[0]])
print('X Shape ', X_cloud_raw.shape)
print('Y Shape ', Y_cloud_raw.shape)
print('Y2 Shape ', Y_cloud2_raw.shape)
print('MinEntries ', minEntries)
```

```python
X_cloud = sorted(random.sample(list(X_cloud_raw),minEntries))
len(X_cloud)
```

```python
Y_cloud = sorted(random.sample(list(Y_cloud_raw),minEntries))
len(Y_cloud)
```

```python
Y_cloud2 = sorted(random.sample(list(Y_cloud2_raw),minEntries))
len(Y_cloud2)
```

```python
import random 
minEntries = min([X_SENSE_raw.shape[0],Y_SENSE_raw.shape[0],Y_SENSE2_raw.shape[0]])
print('X Shape ', X_SENSE_raw.shape)
print('Y Shape ', Y_SENSE_raw.shape)
print('Y2 Shape ', Y_SENSE2_raw.shape)
print('MinEntries ', minEntries)
```

```python
Y_SENSE2 = sorted(random.sample(list(Y_SENSE2_raw),minEntries))
len(Y_SENSE2)
```

```python
X_SENSE2 = sorted(random.sample(list(X_SENSE2_raw),minEntries))
len(X_SENSE2)
```

```python
plt.rcParams.update({'font.size': 15, 'font.family' : "serif"})
fig, ax = plt.subplots(1, 3, figsize=(18, 6))
ax[0].grid(False)
ax[1].grid(False)
ax[2].grid(False)


s = 'Normal: ' + str(auc(X_norm,Y_norm))[:5]
ax[0].plot(X_norm,Y_norm, color = 'black', label= s)

s = '10 m: ' + str(auc(X_RE_2,Y_RE_2))[:5]
ax[0].plot(X_RE_2,Y_RE_2,color='darkgreen',label=s)

s = '20 m: ' + str(auc(X_RE_3,Y_RE_3))[:5]
ax[0].plot(X_RE_3,Y_RE_3,color='forestgreen',label=s)

s = '30 m: ' + str(auc(X_RE_4,Y_RE_4,))[:5]
ax[0].plot(X_RE_4,Y_RE_4,color='limegreen',label=s)

s = '60 m: ' + str(auc(X_RE_5,Y_RE_5))[:5]
ax[0].plot(X_RE_5,Y_RE_5,color='lime',label=s)


ax[0].plot([0,1],[0,1],'--')


ax[0].legend(loc='lower right', shadow=False, ncol=1)
ax[0].set(xlabel="False Positive Rate")
ax[0].set(ylabel="True Positive Rate")
ax[0].set_title("     a)",loc='left')


s = 'Normal: ' + str(auc(X_norm,Y_norm))[:5]
ax[1].plot(X_norm,Y_norm, color = 'black', label= s)

s = 'Gaussian (0.3): ' + str(auc(X_SENSE_raw,Y_SENSE_raw))[:5]
ax[1].plot(X_SENSE_raw,Y_SENSE_raw,color='darkorange', label=s)

s = 'Cloudy (0.7): ' + str(auc(X_cloud2_raw,Y_cloud2_raw))[:5]
ax[1].plot(X_cloud2_raw,Y_cloud2_raw,color='blue',label=s)

s = 'Cloudy (0.6): ' + str(auc(X_cloud_raw,Y_cloud_raw))[:5]
ax[1].plot(X_cloud_raw,Y_cloud_raw, color='cornflowerblue', label=s)

s = 'Gaussian (0.4): ' + str(auc(X_SENSE2_raw,Y_SENSE2_raw))[:5]
ax[1].plot(X_SENSE2_raw,Y_SENSE2_raw,color='orange', label=s)

ax[1].plot([0,1],[0,1],'--')

#ax[1].axes.yaxis.set_visible(False)
ax[1].set_yticklabels(["","","","","",""])

ax[1].legend(loc='lower right', shadow=False, ncol=1)
ax[1].set(xlabel="False Positive Rate")
ax[1].set_title("    b)", loc='left')


s = 'Normal: ' + str(auc(X_norm,Y_norm))[:5]
ax[2].plot(X_norm,Y_norm, color = 'black', label= s)

s = 'Max Diff Site: ' + str(auc(X_CountyMax,Y_CountyMax))[:5]
ax[2].plot(X_CountyMax,Y_CountyMax,color='darkorchid',label=s)

s = 'Avg Diff Site: ' + str(auc(X_DIFF,Y_DIFF))[:5]
ax[2].plot(X_DIFF,Y_DIFF,color='mediumorchid',label=s)

s = 'Min Diff Site: ' + str(auc(X_CountyMin,Y_CountyMin))[:5]
ax[2].plot(X_CountyMin,Y_CountyMin,color='orchid',label=s)

ax[2].plot([0,1],[0,1],'--')

#ax[1].axes.yaxis.set_visible(False)
ax[2].set_yticklabels(["","","","","",""])

ax[2].legend(loc='lower right', shadow=False, ncol=1)
ax[2].set(xlabel="False Positive Rate")
ax[2].set_title("    c)", loc='left')

plt.savefig('Graphs/AUCGraphResampling.png', dpi = 300, transparent = True)
#plt.clf()
plt.show()
```
